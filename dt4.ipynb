```
file_location = "/FileStore/tables/nycdata.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df1)
```

```
df1.printSchema()
```

```
from pyspark.sql.functions import unix_timestamp, from_unixtime , to_timestamp,lit , when , col , sum
df1=df1.select('VendorID',to_timestamp(df1.tpep_pickup_datetime, 'dd-MM-yyyy HH:mm').alias('tpep_pickup_datetime'),to_timestamp(df1.tpep_dropoff_datetime, 'dd-MM-yyyy HH:mm').alias('tpep_dropoff_datetime'),'passenger_count','trip_distance','RatecodeID','store_and_fwd_flag','PULocationID','DOLocationID','payment_type','fare_amount','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount')
```

```
from pyspark.sql.functions import unix_timestamp, from_unixtime , to_timestamp,lit , when , col , sum
from datetime import timedelta
import pyspark.sql.functions as func


maxim = df1.agg({"tpep_pickup_datetime": "max"}).collect()[0]

diff = timedelta(maxim)-func.expr("INTERVAL 10 MINUTES")
df1.filter((df1.tpep_pickup_datetime)>=diff)
```

```
import pyspark.sql.functions as func

df1.groupBy('tpep_pickup_datetime','PULocationID').sum('passenger_count').orderBy('tpep_pickup_datetime','sum(passenger_count)',ascending=False).select(func.col('sum(passenger_count)'),'tpep_pickup_datetime','PULocationID').limit(10).show()
```
